```python
arr1 = np.random.randint(50, 150, 25)
arr1
```

```python
np.argmax(arr1)    ## index 2 = 148 i.e. the max value
```

```python
np.argmin(arr1)   ## index 9 = 50 i.e the min value of this array
```

```python
arr1
```

```python
## Get all the numbers greater than 100

arr1[arr1 > 100]    ## Condition indexing
```

```python
## Get all the numbers greater than 100 and divide those nos by 10

# np.where(condition, name of the array, arithematic operation)

np.where(arr1 > 100, arr1, arr1 / 100)
```

```python
np.where(arr1 % 5 == 0, arr1, arr1 * 5)    ## Check this
```
